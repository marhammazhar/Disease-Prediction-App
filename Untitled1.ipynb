```python
import pandas as pd

# Load dataset
df = pd.read_csv('disease_features.csv')

# Check the first few rows
print(df.head())

# Check column names
print(df.columns.tolist())
```

```python
# Convert lists to space-separated strings
for col in ['Risk Factors', 'Symptoms', 'Signs']:
    df[col] = df[col].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)

# Optional: Display first few rows to confirm
print(df[['Risk Factors', 'Symptoms', 'Signs']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF vectorizers
vectorizer_rf = TfidfVectorizer()
vectorizer_sym = TfidfVectorizer()
vectorizer_signs = TfidfVectorizer()

# Apply TF-IDF to each column
tfidf_rf = vectorizer_rf.fit_transform(df['Risk Factors'])
tfidf_sym = vectorizer_sym.fit_transform(df['Symptoms'])
tfidf_signs = vectorizer_signs.fit_transform(df['Signs'])

# Optional: print shapes
print("TF-IDF Shapes:")
print("Risk Factors:", tfidf_rf.shape)
print("Symptoms    :", tfidf_sym.shape)
print("Signs       :", tfidf_signs.shape)
```

```python
from scipy.sparse import hstack

# Combine the three TF-IDF matrices horizontally
X_tfidf = hstack([tfidf_rf, tfidf_sym, tfidf_signs])

print("Combined TF-IDF Matrix:")
print(f"Shape: {X_tfidf.shape}")
```

```python
import pandas as pd

# Load the one-hot encoded matrix
encoded_df = pd.read_csv('encoded_output2.csv')

# Optional: Check its shape and preview
print("One-Hot Encoded Matrix:")
print(f"Shape: {encoded_df.shape}")
print(encoded_df.head())
```

```python
import numpy as np

# Sparsity = percentage of zero elements
def calculate_sparsity(matrix):
    if hasattr(matrix, "toarray"):  # For sparse matrices
        matrix = matrix.toarray()
    total_elements = matrix.size
    zero_elements = np.count_nonzero(matrix == 0)
    sparsity = (zero_elements / total_elements) * 100
    return sparsity

# Calculate sparsity
sparsity_tfidf = calculate_sparsity(X_tfidf)
sparsity_onehot = calculate_sparsity(encoded_df.values)

print(f"Sparsity of TF-IDF Matrix: {sparsity_tfidf:.2f}%")
print(f"Sparsity of One-Hot Matrix: {sparsity_onehot:.2f}%")
```

```python
# Number of features = number of columns
print(f"TF-IDF Features: {X_tfidf.shape[1]}")
print(f"One-Hot Encoded Features: {encoded_df.shape[1]}")
```

```python
# Drop any non-numeric columns (like labels or IDs)
encoded_numeric = encoded_df.select_dtypes(include=['int64', 'float64'])

# Check the shape
print("Numeric One-Hot Matrix Shape:", encoded_numeric.shape)

# Now apply PCA
from sklearn.decomposition import PCA

pca_onehot = PCA(n_components=3)
pca_onehot_result = pca_onehot.fit_transform(encoded_numeric)

print("PCA on One-Hot Encoding:")
print("Explained Variance Ratio:", pca_onehot.explained_variance_ratio_)
print("Total Explained Variance:", pca_onehot.explained_variance_ratio_.sum())
```

```python
from sklearn.decomposition import TruncatedSVD

# Truncated SVD on one-hot encoded numeric matrix
svd_onehot = TruncatedSVD(n_components=3)
svd_onehot_result = svd_onehot.fit_transform(encoded_numeric)

print("Truncated SVD on One-Hot Encoding:")
print("Explained Variance Ratio:", svd_onehot.explained_variance_ratio_)
print("Total Explained Variance:", svd_onehot.explained_variance_ratio_.sum())
```

```python
# Truncated SVD on TF-IDF matrix
svd_tfidf = TruncatedSVD(n_components=3)
svd_tfidf_result = svd_tfidf.fit_transform(X_tfidf)

print("Truncated SVD on TF-IDF Matrix:")
print("Explained Variance Ratio:", svd_tfidf.explained_variance_ratio_)
print("Total Explained Variance:", svd_tfidf.explained_variance_ratio_.sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_2d_scatter(data_2d, categories, title):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=data_2d[:, 0], y=data_2d[:, 1], hue=categories, palette='Set2', s=70)
    plt.title(title)
    plt.xlabel("Component 1")
    plt.ylabel("Component 2")
    plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics.pairwise import cosine_distances
import numpy as np
import pandas as pd

def evaluate_knn_cv(X, y, k_values, distance_metrics, encoder_name=''):
    results = []

    # Encode string labels
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)

    # 5-fold stratified cross-validation
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    for k in k_values:
        for metric in distance_metrics:
            y_true_all, y_pred_all = [], []

            for train_index, test_index in skf.split(X, y_encoded):
                X_train, X_test = X[train_index], X[test_index]
                y_train, y_test = y_encoded[train_index], y_encoded[test_index]

                try:
                    if metric == 'cosine':
                        clf = KNeighborsClassifier(n_neighbors=k, metric='cosine')
                    else:
                        clf = KNeighborsClassifier(n_neighbors=k, metric=metric)

                    clf.fit(X_train, y_train)
                    y_pred = clf.predict(X_test)

                    y_true_all.extend(y_test)
                    y_pred_all.extend(y_pred)

                except Exception as e:
                    print(f"Skipping k={k}, metric={metric} due to error: {e}")
                    continue

            if len(y_true_all) > 0:
                results.append({
                    'Encoder': encoder_name,
                    'k': k,
                    'Metric': metric,
                    'Accuracy': accuracy_score(y_true_all, y_pred_all),
                    'Precision': precision_score(y_true_all, y_pred_all, average='macro', zero_division=0),
                    'Recall': recall_score(y_true_all, y_pred_all, average='macro', zero_division=0),
                    'F1-score': f1_score(y_true_all, y_pred_all, average='macro', zero_division=0)
                })

    return pd.DataFrame(results)
```

```python
# Use smaller k-values first
k_vals = [1, 3]
distances = ['euclidean', 'manhattan', 'cosine']

results_tfidf = evaluate_knn_cv(X_tfidf, y, k_vals, distances, encoder_name='TF-IDF')
results_onehot = evaluate_knn_cv(X_onehot, y, k_vals, distances, encoder_name='One-Hot')

final_results = pd.concat([results_tfidf, results_onehot], ignore_index=True)
final_results
```

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd

def evaluate_knn_loocv(X, y, k_values, distance_metrics, encoder_name=''):
    results = []
    
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)

    loo = LeaveOneOut()

    for k in k_values:
        for metric in distance_metrics:
            y_true_all, y_pred_all = [], []

            for train_index, test_index in loo.split(X):
                X_train, X_test = X[train_index], X[test_index]
                y_train, y_test = y_encoded[train_index], y_encoded[test_index]

                clf = KNeighborsClassifier(n_neighbors=k, metric=metric)
                try:
                    clf.fit(X_train, y_train)
                    y_pred = clf.predict(X_test)
                    y_true_all.extend(y_test)
                    y_pred_all.extend(y_pred)
                except Exception as e:
                    print(f"Skipping k={k}, metric={metric} due to error: {e}")
                    break

            if y_true_all and y_pred_all:
                results.append({
                    'Encoder': encoder_name,
                    'k': k,
                    'Metric': metric,
                    'Accuracy': accuracy_score(y_true_all, y_pred_all),
                    'Precision': precision_score(y_true_all, y_pred_all, average='macro', zero_division=0),
                    'Recall': recall_score(y_true_all, y_pred_all, average='macro', zero_division=0),
                    'F1-score': f1_score(y_true_all, y_pred_all, average='macro', zero_division=0)
                })

    return pd.DataFrame(results)
```

```python
k_vals = [1, 3, 5]
distances = ['euclidean', 'manhattan', 'cosine']

results_tfidf = evaluate_knn_loocv(X_tfidf.toarray(), y, k_vals, distances, encoder_name='TF-IDF')
results_onehot = evaluate_knn_loocv(X_onehot, y, k_vals, distances, encoder_name='One-Hot')

final_results = pd.concat([results_tfidf, results_onehot], ignore_index=True)
final_results
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_logreg_loocv(X, y, encoder_name=''):
    results = []
    
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)

    loo = LeaveOneOut()
    y_true_all, y_pred_all = [], []

    for train_index, test_index in loo.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y_encoded[train_index], y_encoded[test_index]

        clf = LogisticRegression(max_iter=1000)
        try:
            clf.fit(X_train, y_train)
            y_pred = clf.predict(X_test)
            y_true_all.extend(y_test)
            y_pred_all.extend(y_pred)
        except Exception as e:
            print(f"Error in Logistic Regression LOOCV: {e}")
            break

    if y_true_all and y_pred_all:
        results.append({
            'Encoder': encoder_name,
            'Model': 'LogisticRegression',
            'Accuracy': accuracy_score(y_true_all, y_pred_all),
            'Precision': precision_score(y_true_all, y_pred_all, average='macro', zero_division=0),
            'Recall': recall_score(y_true_all, y_pred_all, average='macro', zero_division=0),
            'F1-score': f1_score(y_true_all, y_pred_all, average='macro', zero_division=0)
        })

    return pd.DataFrame(results)
```

```python
results_logreg_tfidf = evaluate_logreg_loocv(X_tfidf.toarray(), y, encoder_name='TF-IDF')
results_logreg_onehot = evaluate_logreg_loocv(X_onehot, y, encoder_name='One-Hot')
```

```python
results_logreg = pd.concat([results_logreg_tfidf, results_logreg_onehot], ignore_index=True)
results_logreg['k'] = '-'  # No k for Logistic Regression
results_logreg['Metric'] = '-'  # No distance metric
results_logreg['Model'] = 'LogisticRegression'

# Add model label to KNN results
final_results['Model'] = 'KNN'

# Combine both
all_results = pd.concat([final_results, results_logreg], ignore_index=True)
all_results = all_results[['Model', 'Encoder', 'k', 'Metric', 'Accuracy', 'Precision', 'Recall', 'F1-score']]
all_results
```
